## Bar Plot - Year vs No. Of Boundaries

```python
def plot_boundaries(data, format_name, color):
    # Calculate the total number of boundaries (4s and 6s) for each year
    boundaries_data = data.groupby("Year")[["4s", "6s"]].sum()
    boundaries_data["Total"] = boundaries_data["4s"] + boundaries_data["6s"]

    # Create a bar plot of the number of boundaries each year
    plt.figure(figsize=(5, 4))
    plt.bar(boundaries_data.index, boundaries_data["Total"], color=color)
    plt.title(f"{format_name} - Year vs Number of Boundaries")
    plt.xlabel("Year")
    plt.ylabel("Number of Boundaries")
    plt.show()

# Define a list of colors to use for each plot
colors = ["red", "green", "blue", "purple"]

# Call the plot_boundaries function for each format with a different color
plot_boundaries(odi_data, "ODI", colors[0])
plot_boundaries(t20_data, "T20", colors[1])
plot_boundaries(test_data, "Test", colors[2])
plot_boundaries(ipl_data, "IPL", colors[3])
```

## Total Number of 4s and 6s Scored by Rohit Sharma in each format

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
odi_data = pd.read_csv('ODI.csv')
t20_data = pd.read_csv('T20.csv')
test_data = pd.read_csv('TEST.csv')
ipl_data = pd.read_csv('IPL.csv')

# Define a function to display the bar chart for the selected format
def display_bar_chart(format):
    # Get the data for the selected format
    data = {"ODI": odi_data, "T20": t20_data, "Test": test_data, "IPL": ipl_data}[format]

    # Calculate the total number of 4s and 6s for each year
    runs_by_year = data.groupby('Year')[['4s', '6s']].sum()


    # Define the colors for the bars based on the format
    colors = ['#0050A0', '#FFA500'] if format == 'ODI' else \
             ['#FF4500', '#32CD32'] if format == 'T20' else \
             ['#800000', '#FF8C00'] if format == 'Test' else \
             ['#00BFFF', '#DC143C']

    # Create a bar chart of the total number of 4s and 6s scored by year
    ax = runs_by_year.plot(kind='bar', color=colors)
    ax.set_xlabel("Year")
    ax.set_ylabel("Number of Runs")
    ax.set_title(f"Total Number of 4s and 6s Scored by Rohit Sharma in {format}")

    # Display the plot
    plt.show()

# Call the function for each format to display the bar chart
formats = ['ODI', 'T20', 'Test', 'IPL']
for format in formats:
    display_bar_chart(format)
```

## GUI- Half-Centuries and Centuries over time of Rohit Sharma

## GUI- Time series Analysis

```python
from statsmodels.tsa.seasonal import seasonal_decompose

#Moving average of Runs in ODI matches by year
odi_data['Runs'] = pd.to_numeric(odi_data['Runs'], errors='coerce')
odi_data.groupby('Year')['Runs'].mean().rolling(window=3).mean().plot()
plt.title("ODI - Moving average of Runs by Year")
plt.xlabel("Year")
plt.ylabel("Runs")
plt.show()

#Exponential smoothing of Avg in T20 matches by year
t20_data['Avg'] = pd.to_numeric(t20_data['Avg'], errors='coerce')
t20_data.groupby('Year')['Avg'].mean().ewm(span=2).mean().plot()
plt.title("T20 - Exponential smoothing of Avg by Year")
plt.xlabel("Year")
plt.ylabel("Avg")
plt.show()

#Seasonal decomposition of SR in TEST matches
decomp = seasonal_decompose(test_data['SR'], period=5)
decomp.plot()
plt.suptitle("TEST - Seasonal decomposition of SR")
plt.show()

#Autocorrelation of 6s in IPL matches
pd.plotting.autocorrelation_plot(ipl_data['6s'])
plt.title("IPL - Autocorrelation of 6s")
plt.show()
```

## GUI FOR DISPLAYING FULL STATS FOR EACH FORMAT

## GUI - Box Plots for each format

```python
import tkinter as tk
from tkinter import ttk
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
odi_data = pd.read_csv('ODI.csv')
t20_data = pd.read_csv('T20.csv')
test_data = pd.read_csv('TEST.csv')
ipl_data = pd.read_csv('IPL.csv')
# Create a dictionary of formats and their corresponding dataframes
formats = {"ODI": odi_data, "T20": t20_data, "Test": test_data, "IPL": ipl_data}

# Define a function to create the box plots for a given format
def create_box_plots(format_name, df):
    plt.figure(figsize=(10, 6))
    for j, column in enumerate(df.columns[2:]):
        data = pd.to_numeric(df[column], errors='coerce').dropna().values
        plt.subplot(3, 4, j+1)
        plt.boxplot(data)
        plt.title(column)
        plt.ylabel(column)
        plt.xticks([])
    plt.suptitle(f"{format_name} - Box Plot")
    plt.show()

# Define a function to be called when a button is clicked
def button_click(format_name):
    df = formats[format_name]
    create_box_plots(format_name, df)

# Create the main window
root = tk.Tk()
root.title("Rohit Sharma Performance Analysis - Box Plot")

# Create a label for the window
label = ttk.Label(root, text="Select a format to view the box plots:")
label.pack()

# Create a button for each format
for format_name in formats.keys():
    button = tk.Button(root, text=format_name, command=lambda f=format_name: button_click(f))
    button.pack(side=tk.TOP, anchor=tk.CENTER)

# Run the GUI
root.mainloop()
```

```python
import tkinter as tk
import pandas as pd
import statsmodels.formula.api as smf

# Load data
odi_data = pd.read_csv("ODI.csv")
test_data = pd.read_csv("TEST.csv")
t20_data = pd.read_csv("T20.csv")
ipl_data = pd.read_csv("IPL.csv")

# Create main window
root = tk.Tk()
root.title("Linear Regression Analysis")
root.state('zoomed')

# Define function to run linear regression and show results
def run_linear_regression(format_name, data):
    model = smf.ols('Runs ~ Avg', data=data).fit()
    output_text.config(state=tk.NORMAL)
    output_text.delete("1.0", tk.END)
    output_text.insert(tk.END, f"{format_name} - Linear Regression Results\n\n{model.summary()}")
    output_text.config(state=tk.DISABLED)

# Define function to handle button click
def button_click(format_name, data):
    run_linear_regression(format_name, data)

# Create buttons
odi_button = tk.Button(root, text="ODI", command=lambda: button_click("ODI", odi_data))
test_button = tk.Button(root, text="Test", command=lambda: button_click("Test", test_data))
t20_button = tk.Button(root, text="T20", command=lambda: button_click("T20", t20_data))
ipl_button = tk.Button(root, text="IPL", command=lambda: button_click("IPL", ipl_data))

# Pack buttons
odi_button.pack(side=tk.TOP, padx=0, pady=0)
test_button.pack(side=tk.TOP, padx=0, pady=0)
t20_button.pack(side=tk.TOP, padx=0, pady=0)
ipl_button.pack(side=tk.TOP, padx=0, pady=0)

# Create output window
output_frame = tk.Frame(root)
output_frame.pack(fill=tk.BOTH, expand=True)

output_text = tk.Text(output_frame, state=tk.DISABLED)
output_text.pack(fill=tk.BOTH, expand=True)

root.mainloop()
```

```python
import tkinter as tk
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Load data
odi_data = pd.read_csv("ODI.csv")
test_data = pd.read_csv("TEST.csv")
t20_data = pd.read_csv("T20.csv")
ipl_data = pd.read_csv("IPL.csv")

# Create main window
root = tk.Tk()
root.title("Time Analysis")

# Define function to show ODI analysis
def show_odi_analysis():
    #Moving average of Runs in ODI matches by year
    odi_data['Runs'] = pd.to_numeric(odi_data['Runs'], errors='coerce')
    odi_data.groupby('Year')['Runs'].mean().rolling(window=3).mean().plot()
    plt.title("ODI - Moving average of Runs by Year")
    plt.xlabel("Year")
    plt.ylabel("Runs")
    plt.show()

# Define function to show T20 analysis
def show_t20_analysis():
    #Exponential smoothing of Avg in T20 matches by year
    t20_data['Avg'] = pd.to_numeric(t20_data['Avg'], errors='coerce')
    t20_data.groupby('Year')['Avg'].mean().ewm(span=2).mean().plot()
    plt.title("T20 - Exponential smoothing of Avg by Year")
    plt.xlabel("Year")
    plt.ylabel("Avg")
    plt.show()

# Define function to show Test analysis
def show_test_analysis():
    #Seasonal decomposition of SR in TEST matches
    decomp = seasonal_decompose(test_data['SR'], period=5)
    decomp.plot()
    plt.suptitle("TEST - Seasonal decomposition of SR")
    plt.show()

# Define function to show IPL analysis
def show_ipl_analysis():
    #Autocorrelation of 6s in IPL matches
    pd.plotting.autocorrelation_plot(ipl_data['6s'])
    plt.title("IPL - Autocorrelation of 6s")
    plt.show()

# Create buttons
odi_button = tk.Button(root, text="ODI", command=show_odi_analysis)
test_button = tk.Button(root, text="Test", command=show_test_analysis)
t20_button = tk.Button(root, text="T20", command=show_t20_analysis)
ipl_button = tk.Button(root, text="IPL", command=show_ipl_analysis)

# Pack buttons
odi_button.pack(side=tk.TOP, padx=0, pady=0)
test_button.pack(side=tk.TOP, padx=0, pady=0)
t20_button.pack(side=tk.TOP, padx=0, pady=0)
ipl_button.pack(side=tk.TOP, padx=0, pady=0)
root.mainloop()
```

## GUI - Pie Chart Analysis

```python
from statsmodels.tsa.seasonal import seasonal_decompose
import pandas as pd
import matplotlib.pyplot as plt
import tkinter as tk

# Clean the data by converting appropriate columns to numeric data type and filling missing values
odi_data['Runs'] = pd.to_numeric(odi_data['Runs'], errors='coerce')
t20_data['Avg'] = pd.to_numeric(t20_data['Avg'], errors='coerce')
test_data['SR'] = pd.to_numeric(test_data['SR'], errors='coerce')
ipl_data['4s'] = pd.to_numeric(ipl_data['4s'], errors='coerce')
ipl_data['6s'] = pd.to_numeric(ipl_data['6s'], errors='coerce')

# Create a tkinter window
root = tk.Tk()

# Define functions to create pie charts and display them in the window
def show_odi_pie():
    avg_runs = odi_data.groupby('Innings')['Runs'].mean()
    plt.pie(avg_runs, labels=avg_runs.index, autopct='%1.1f%%')
    plt.title("Average Runs Scored by Rohit Sharma in ODI Innings")
    plt.show()

def show_t20_pie():
    outs_pct = t20_data.groupby('Year')['Outs'].sum() / t20_data.groupby('Year')['Innings'].count()
    plt.pie(outs_pct, labels=outs_pct.index, autopct='%1.1f%%')
    plt.title("Percentage of Outs by Rohit Sharma in T20 Matches")
    plt.show()

def show_test_pie():
    boundaries_pct = test_data.groupby('Year')['4s', '6s'].sum().sum(axis=1) / test_data.groupby('Year')['Balls'].count()
    plt.pie(boundaries_pct, labels=boundaries_pct.index, autopct='%1.1f%%')
    plt.title("Percentage of Boundaries Hit by Rohit Sharma in Test Matches")
    plt.show()

def show_ipl_pie():
    avg_pct = ipl_data.groupby('Year')['Avg'].mean()
    plt.pie(avg_pct, labels=avg_pct.index, autopct='%1.1f%%')
    plt.title("Average runs Percentage by Rohit Sharma in IPL Matches")
    plt.show()

# Create buttons to display the pie charts
odi_button = tk.Button(root, text='ODI', command=show_odi_pie)
t20_button = tk.Button(root, text='T20', command=show_t20_pie)
test_button = tk.Button(root, text='Test', command=show_test_pie)
ipl_button = tk.Button(root, text='IPL', command=show_ipl_pie)

# Add buttons to the window
odi_button.pack(side=tk.LEFT, padx=10, pady=10)
t20_button.pack(side=tk.LEFT, padx=10, pady=10)
test_button.pack(side=tk.LEFT, padx=10, pady=10)
ipl_button.pack(side=tk.LEFT, padx=10, pady=10)

# Run the tkinter event loop
root.mainloop()
```

##  GUI - Heatmap for each format

```python
import tkinter as tk
from tkinter import filedialog
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Define a function to read the data from a csv file and return a pandas DataFrame
def read_data(file_path):
    return pd.read_csv(file_path)


# Define a function to display the heat map of correlation matrix for a given DataFrame
def display_heatmap(data, title):
    plt.figure(figsize=(7, 5))
    sns.heatmap(data.corr(), cmap='Blues', annot=True)
    plt.title(title)
    plt.show()


# Create the main window
root = tk.Tk()
root.title("Heat Maps")

# Define a function to open a file dialog and get the file path of a csv file
def open_file():
    file_path = filedialog.askopenfilename(filetypes=[('CSV files', '*.csv')])
    if file_path:
        data = read_data(file_path)
        title = f"Correlation matrix - {file_path.split('/')[-1].split('.')[0]}"
        display_heatmap(data, title)

# Create buttons for each format
button_odi = tk.Button(root, text="ODI", command=lambda: display_heatmap(read_data('odi.csv'), "Correlation matrix - ODI"))
button_t20 = tk.Button(root, text="T20", command=lambda: display_heatmap(read_data('t20.csv'), "Correlation matrix - T20"))
button_test = tk.Button(root, text="Test", command=lambda: display_heatmap(read_data('test.csv'), "Correlation matrix - Test"))
button_ipl = tk.Button(root, text="IPL", command=lambda: display_heatmap(read_data('ipl.csv'), "Correlation matrix - IPL"))

# Add buttons to the main window
button_odi.pack(side=tk.LEFT, padx=10)
button_t20.pack(side=tk.LEFT, padx=10)
button_test.pack(side=tk.LEFT, padx=10)
button_ipl.pack(side=tk.LEFT, padx=10)

# Start the main loop
root.mainloop()
```
